```python
# Customer segmentation Visualisation
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create tenure groups
bins = [0, 12, 24, 48, 72]
labels = ['0-12 Months', '13-24 Months', '25-48 Months', '49-72 Months']

df['tenure_group'] = pd.cut(
    df['tenure'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Count customers in each tenure group
tenure_group = df['tenure_group'].value_counts().sort_index()

# Create doughnut chart
plt.figure(figsize=(8, 8))

plt.pie(
    tenure_group,
    labels=tenure_group.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width': 0.4, 'edgecolor': 'white'},
    pctdistance=0.8
)

# Draw a white circle in the center
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
plt.gca().add_artist(centre_circle)

plt.title('Customer Distribution by Tenure Group')
plt.tight_layout()
plt.show()
```

```python
#Create Tenure Groups
import pandas as pd
import matplotlib.pyplot as plt

bins = [0, 12, 36, 72]
labels = ['0-12 Months', '13-36 Months', '37+ Months']

df['tenure_group'] = pd.cut(
    df['tenure'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
#Calculate the average monthly charges

avg_monthly = df.groupby('tenure_group')['monthlycharges'].mean()

avg_monthly
```

```python
cluster = df.groupby(['tenure_group', 'churn'])['monthlycharges'].mean().unstack()

cluster.plot(
    kind='bar',
    figsize=(9,5)
)

plt.title('Average Monthly Charges by Tenure Group and Churn')
plt.xlabel('Tenure Group')
plt.ylabel('Average Monthly Charges')
plt.xticks(rotation=0)

plt.legend(title='Churn')

plt.show()
```

```python
#Add Annotations to Highlight Significant Trends
ax = cluster.plot(
    kind='bar',
    figsize=(9,5)
)

plt.title('Average Monthly Charges by Tenure Group and Churn')
plt.xlabel('Tenure Group')
plt.ylabel('Average Monthly Charges')
plt.xticks(rotation=0)

# Add value labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)

plt.show()
```

```python
#Add a Business Insight Annotation
ax = cluster.plot(kind='bar', figsize=(9,5))

plt.title('Average Monthly Charges by Tenure Group and Churn')
plt.xlabel('Tenure Group')
plt.ylabel('Average Monthly Charges')

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f')

plt.annotate(
    'Customers with higher monthly\ncharges tend to churn more.',
    xy=(1, cluster.iloc[1,1]),
    xytext=(1.5, cluster.values.max()+5),
    arrowprops=dict(arrowstyle='->')
)

plt.show()
```

```python
#Group Customers by Tenure to Compute Statistics for Charges and Churn
```

```python
#Create tenure groups

bins = [0, 12, 36, 72]
labels = ['0-12 Months', '13-36 Months', '37+ Months']

df['tenure_group'] = pd.cut(
    df['tenure'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
#Compute statistics

group_stats = df.groupby('tenure_group').agg({
    'monthlycharges': ['mean', 'median', 'min', 'max'],
    'totalcharges': ['mean', 'median', 'min', 'max']
})

group_stats
```

```python
#Churn count by tenure group
pd.crosstab(df['tenure_group'], df['churn'])
```

```python
#Churn percentage by tenure group
round(
    pd.crosstab(
        df['tenure_group'],
        df['churn'],
        normalize='index'
    ) * 100,
    2
)
```

```python
#Analyze Churn Rates by Demographics
```

```python
#Gender
gender_churn = round(
    pd.crosstab(
        df['gender'],
        df['churn'],
        normalize='index'
    ) * 100,
    2
)

gender_churn
```

```python
#Bar Chart
gender_churn.plot(kind='bar', figsize=(7,5))

plt.title('Churn Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.xticks(rotation=0)

plt.show()
```

```python
#Senior Citizen
senior_churn = round(
    pd.crosstab(
        df['seniorcitizen'],
        df['churn'],
        normalize='index'
    ) * 100,
    2
)

senior_churn
```

```python
#Bar Chart
senior_churn.plot(kind='bar', figsize=(7,5))

plt.title('Churn Rate by Senior Citizen')
plt.xlabel('Senior Citizen')
plt.ylabel('Percentage')
plt.xticks([0,1], ['No','Yes'], rotation=0)

plt.show()
```

```python
#Churn Rates by Payment Methods
payment_churn = round(
    pd.crosstab(
        df['paymentmethod'],
        df['churn'],
        normalize='index'
    ) * 100,
    2
)

payment_churn
```

```python
#Chart
payment_churn.plot(kind='bar', figsize=(9,5))

plt.title('Churn Rate by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Percentage')
plt.xticks(rotation=20)

plt.show()
```

```python
#Churn Rates by Contract Type
contract_churn = round(
    pd.crosstab(
        df['contract'],
        df['churn'],
        normalize='index'
    ) * 100,
    2
)

contract_churn
```

```python
#Chart
contract_churn.plot(kind='bar', figsize=(8,5))

plt.title('Churn Rate by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Percentage')

plt.xticks(rotation=0)

plt.show()
```

```python
#Visualize Trends Over Customer Lifecycle
lifecycle = round(
    pd.crosstab(
        df['tenure_group'],
        df['churn'],
        normalize='index'
    ) * 100,
    2
)

lifecycle
```

```python
#Line Chart
lifecycle.plot(
    kind='line',
    marker='o',
    figsize=(8,5)
)

plt.title('Customer Churn Across Lifecycle Stages')
plt.xlabel('Tenure Group')
plt.ylabel('Percentage')

plt.grid(True)

plt.show()
```
